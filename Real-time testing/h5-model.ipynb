```python
import cv2
#import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
input_shape = (256, 256, 3)
n_classes = 1
model = DeepLabV3Plus(inputs=input_shape, classes=n_classes)
model.load_weights('PipeSeg_30.h5')
```

```python
cap = cv2.VideoCapture(0)
```

```python
while(True):
    ret, frame = cap.read()
    #frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY )
    img = cv2.resize(frame, (256, 256)).astype(np.float32)/255.0
    pred = model.predict(img.reshape(1, 256, 256, 3))
    pred = np.where(pred>0.5, 1, 0).reshape(256, 256, 1)
    
    pred = cv2.bitwise_not(pred)
    pred = np.where(pred==-2, 0.0, 1.0)
    #img = cv2.resize(img, (256,256,3))
    #pred = cv2.resize(pred,(256,256,3))
    #vis = np.concatenate((img, pred), axis=0)
    cv2.imshow('org',img)
    cv2.imshow('pred',pred)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
